```python
# import modules
import sklearn as sk
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read csv
df = pd.read_csv("laptopPrice.csv")
df.head()
```

```python
# check nun value
df.isnull().sum()
```

```python
df.info()
```

```python
# plot graphs
sns.pairplot(df)
plt.show()
```

```python
# split object value
cat_val = df.select_dtypes(include="O")
cat_val
```

```python
# split numerical value
num_val = df.select_dtypes(exclude="O")
num_val
```

```python
for i in cat_val:
    print(i ,"->", cat_val[i].unique())
```

```python
# label encoding 
l = ['brand','processor_name','ram_type','rating','warranty','ssd','hdd','processor_gnrtn']
for i in l:
    d = {j:key for key , j in enumerate(list(cat_val[i].unique()))}
    cat_val[i] = cat_val[i].map(d)
cat_val
```

```python
# one hot encoding 
cat_val = pd.get_dummies(cat_val,columns=["processor_brand","ram_gb","os","os_bit","graphic_card_gb","weight","Touchscreen","msoffice"],drop_first=True)
```

```python
# make a final dataframe
final_df = pd.concat([cat_val,num_val],axis=1)
```

```python
final_df
```

```python
# split features and labels
X = final_df.drop("Price",axis=1)
Y = final_df.Price
```

```python
# split train data 80% and test data 20%
xtrain,xtest,ytrain,ytest = sk.model_selection.train_test_split(X,Y,test_size=0.2,random_state = 42)
```

```python
# load model
model = sk.linear_model.LinearRegression()
```

```python
# fit model on train dataset
model.fit(xtrain,ytrain)
```

```python
# check score
model.score(xtrain,ytrain),model.score(xtest,ytest)
```

```python
# plot the best fit line
plt.scatter(ytest,model.predict(xtest))
plt.plot(ytest,ytest)
plt.show()
```
